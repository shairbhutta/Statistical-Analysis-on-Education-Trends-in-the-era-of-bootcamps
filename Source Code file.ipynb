**Descriptive Statistics**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
response_file_path = r'C:\Users\AA\Desktop\Survey Data File.csv'
response_data = pd.read_csv(response_file_path)

# Summary statistics for numerical columns
numerical_summary = response_data.describe()

# Frequency distribution for categorical columns
categorical_summary = {}
for column in response_data.select_dtypes(include=['object']):
    categorical_summary[column] = response_data[column].value_counts()

# Display numerical summary
print("Numerical Summary:")
display(numerical_summary)  # In Jupyter, this will render as a table

# Visualization for numerical columns
numerical_columns = response_data.select_dtypes(include=['number']).columns
for column in numerical_columns:
    plt.figure(figsize=(8, 6))
    response_data[column].dropna().hist(bins=15, alpha=0.75, color='blue', edgecolor='black')
    plt.title(f"Histogram of {column}")
    plt.xlabel(column)
    plt.ylabel("Frequency")
    plt.grid(False)
    plt.show()

# Display categorical summary
print("\nCategorical Summary:")
for col, freq in categorical_summary.items():
    print(f"\nColumn: {col}")
    display(freq)  # In Jupyter, this will render each frequency distribution as a table

    # Visualization for categorical columns
    plt.figure(figsize=(10, 6))
    freq.plot(kind='bar', color='skyblue', edgecolor='black')
    plt.title(f"Bar Chart of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.xticks(rotation=45, ha='right')
    plt.show()
```

**Tests**

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency

# Preparing a contingency table
contingency_table = pd.crosstab(
    response_data['What is your gender? '],
    response_data['How likely are you to consider enrolling in a bootcamp or online certification in the future? ']
)

# Performing the Chi-Square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Organizing results
chi_square_results = pd.DataFrame({
    "Statistic": ["Chi-Square Statistic", "P-Value", "Degrees of Freedom"],
    "Value": [chi2, p, dof]
})

# Displaying the contingency table
print("Contingency Table (Gender vs Bootcamp Enrollment Likelihood):")
display(contingency_table)  # This will render as a table in Jupyter

# Visualization for contingency table
contingency_table.plot(kind="bar", stacked=True, figsize=(10, 6), color=['skyblue', 'orange', 'green', 'red', 'purple'])
plt.title("Bootcamp Enrollment Likelihood by Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Responses")
plt.legend(title="Enrollment Likelihood")
plt.xticks(rotation=0)
plt.show()

# Displaying the Chi-Square Test Results
print("\nChi-Square Test Results:")
display(chi_square_results)  # This will render as a table in Jupyter

# Adding a visualization for expected frequencies
expected_df = pd.DataFrame(expected, 
                           index=contingency_table.index, 
                           columns=contingency_table.columns)

# Plot expected frequencies
expected_df.plot(kind="bar", stacked=True, figsize=(10, 6), color=['lightblue', 'lightcoral', 'lightgreen', 'lightpink', 'lightgrey'])
plt.title("Expected Frequencies for Bootcamp Enrollment by Gender")
plt.xlabel("Gender")
plt.ylabel("Expected Number of Responses")
plt.legend(title="Enrollment Likelihood")
plt.xticks(rotation=0)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency

# Perform Chi-Square test for employer-related variables
# Variables: 'Do you think employers value bootcamps and online certifications as much as traditional degrees?'
#            vs. 'How likely are employers in your field to consider a bootcamp graduate equally with a university degree holder?'

# Preparing a contingency table
employer_contingency_table = pd.crosstab(
    response_data['Do you think employers value bootcamps and online certifications as much as traditional degrees? '],
    response_data['How likely are employers in your field to consider a bootcamp graduate equally with a university degree holder? ']
)

# Performing the Chi-Square test
chi2, p, dof, expected = chi2_contingency(employer_contingency_table)

# Organizing results
employer_chi_square_results = pd.DataFrame({
    "Statistic": ["Chi-Square Statistic", "P-Value", "Degrees of Freedom"],
    "Value": [chi2, p, dof]
})

# Displaying the contingency table
print("Contingency Table (Updated Employer Perspective):")
display(employer_contingency_table)  # This will render the table in Jupyter

# Visualization for contingency table
employer_contingency_table.plot(kind="bar", figsize=(12, 6), color=['skyblue', 'orange', 'green', 'red', 'purple'])
plt.title("Employer Valuation of Bootcamps vs Likelihood to Hire")
plt.xlabel("Employer Valuation")
plt.ylabel("Number of Responses")
plt.legend(title="Likelihood to Hire")
plt.xticks(rotation=45, ha='right')
plt.show()

# Displaying the Chi-Square Test Results
print("\nUpdated Chi-Square Test Results (Employer Perspective):")
display(employer_chi_square_results)  # This will render the results as a table in Jupyter

# Adding a visualization for expected frequencies
expected_df = pd.DataFrame(expected, 
                           index=employer_contingency_table.index, 
                           columns=employer_contingency_table.columns)

# Plot expected frequencies
expected_df.plot(kind="bar", figsize=(12, 6), color=['lightblue', 'lightcoral', 'lightgreen', 'lightpink', 'lightgrey'])
plt.title("Expected Frequencies for Employer Perspective")
plt.xlabel("Employer Valuation")
plt.ylabel("Expected Number of Responses")
plt.legend(title="Likelihood to Hire")
plt.xticks(rotation=45, ha='right')
plt.show()
```

****ANOVA****

```python
from scipy.stats import f_oneway
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Extract relevant columns for ANOVA
education_levels = response_data['What is your current level of education? ']
skill_usefulness_ratings = response_data[
    'On a scale of 1-5, how would you rate the usefulness of the skills gained from your alternative education pathway (e.g., bootcamp, online certification) for your current job? '
]

# Group skill usefulness ratings by education levels and filter out groups with fewer than 5 samples
anova_groups = [
    skill_usefulness_ratings[education_levels == level].dropna()
    for level in education_levels.unique()
]

# Filter out groups with less than 5 samples
anova_groups = [group for group in anova_groups if len(group) >= 5]

if len(anova_groups) > 1:  # Ensure we have at least two groups for ANOVA
    # Perform ANOVA test
    f_stat, p_value = f_oneway(*anova_groups)

    # Organize results
    anova_results = pd.DataFrame({
        "Statistic": ["F-Statistic", "P-Value"],
        "Value": [f_stat, p_value]
    })

    # Display the results
    print("ANOVA Test Results (Usefulness of Bootcamps Across Education Levels):")
    print(anova_results)

    # Visualization: Boxplot for skill usefulness by education level
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=education_levels, y=skill_usefulness_ratings)
    plt.title("Usefulness of Bootcamps Across Education Levels")
    plt.xlabel("Education Levels")
    plt.ylabel("Skill Usefulness Ratings (1-5)")
    plt.xticks(rotation=45, ha='right')
    plt.show()

else:
    print("Not enough valid groups for ANOVA test.")
```

```python
from scipy.stats import kruskal
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Clean data: Filter valid education levels
valid_education_levels = ['Under-Graduate', 'Post-Graduate', "Bachelor's Degree", "Master's Degree", "Doctoral Degree"]
filtered_data = response_data[
    response_data['What is your current level of education? '].isin(valid_education_levels)
]

# Extract relevant columns
education_levels = filtered_data['What is your current level of education? ']
investment_willingness = filtered_data[
    'How much more willing are you to invest in alternative education pathways (e.g., bootcamps, online certifications) compared to traditional higher education? '
].dropna()

# Group data by education levels
kruskal_groups = [
    investment_willingness[education_levels == level]
    for level in valid_education_levels
]

# Check group sizes after filtering
for level, group in zip(valid_education_levels, kruskal_groups):
    print(f"Group for {level} has {len(group)} data points")

# Perform Kruskal-Wallis test
kruskal_stat, kruskal_p = kruskal(*kruskal_groups)

# Display the results
print("\nKruskal-Wallis Test")
print(f"Statistic: {kruskal_stat}, P-Value: {kruskal_p}")

# Visualization: Boxplot for investment willingness by education level
plt.figure(figsize=(10, 6))
sns.boxplot(
    x=filtered_data['What is your current level of education? '],
    y=filtered_data[
        'How much more willing are you to invest in alternative education pathways (e.g., bootcamps, online certifications) compared to traditional higher education? '
    ]
)
plt.title("Investment Willingness by Education Level")
plt.xlabel("Education Level")
plt.ylabel("Investment Willingness (1-5)")
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
from scipy.stats import spearmanr
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Convert categorical responses to numerical codes
importance_traditional_degrees = response_data[
    'How important do you believe traditional degrees are for career success in your field? '
].astype('category').cat.codes

effectiveness_bootcamps = response_data[
    'How effective do you believe bootcamps and online certifications are for career advancement compared to traditional degrees? '
].astype('category').cat.codes

# Compute Spearman rank correlation
rank_corr_coeff, rank_corr_p = spearmanr(importance_traditional_degrees, effectiveness_bootcamps)
print("\nRank Correlation")
print(f"Coefficient: {rank_corr_coeff}")
print(f"P-Value: {rank_corr_p}")

# Visualization: Scatter plot with regression line
plt.figure(figsize=(10, 6))
sns.regplot(
    x=importance_traditional_degrees,
    y=effectiveness_bootcamps,
    ci=None,
    scatter_kws={"alpha": 0.7},
    line_kws={"color": "red"}
)
plt.title("Importance of Traditional Degrees vs Effectiveness of Bootcamps")
plt.xlabel("Importance of Traditional Degrees (Encoded)")
plt.ylabel("Effectiveness of Bootcamps (Encoded)")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for the bar chart
categories = [
    "Usefulness of Bootcamps Across Education Levels",
    "Usefulness of Skills Ratings Across Employment Status"
]
f_stat_values = [2.13, 0.873]
p_values = [0.0650, 0.5000]

# Colors based on significance
colors = ['green' if p < 0.05 else 'darkgreen' for p in p_values]

# Create the horizontal bar chart
fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.barh(categories, f_stat_values, color=colors, alpha=0.8)

# Add annotations for the F-statistics and p-values
for bar, p_val, f_stat in zip(bars, p_values, f_stat_values):
    ax.text(bar.get_width() + 0.1, bar.get_y() + bar.get_height() / 2,
            f"$F$ = {f_stat:.2f}, p = {p_val:.4f}",
            va='center', fontsize=10)

# Add title and labels
ax.set_title("ANOVA Results", fontsize=14, weight='bold')
ax.set_xlabel("F-Statistic Value", fontsize=12)
ax.set_xlim(0, max(f_stat_values) + 1)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for the bar chart
categories = [
    "Gender vs Enrollment in Bootcamp",
    "Employer Valuation vs Hiring Likelihood",
    "Education Level vs Likelihood of Bootcamp Enrollment"
]
chi_square_values = [2.82, 49.32, 12.45]
p_values = [0.5900, 0.0001, 0.0050]

# Colors based on significance
colors = ['red' if p > 0.05 else 'green' for p in p_values]

# Create the horizontal bar chart
fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.barh(categories, chi_square_values, color=colors, alpha=0.8)

# Add annotations for the p-values and Chi-Square statistics
for bar, p_val, chi_sq in zip(bars, p_values, chi_square_values):
    ax.text(bar.get_width() + 1, bar.get_y() + bar.get_height() / 2,
            f"$\\chi^2$ = {chi_sq:.2f}, p = {p_val:.4f}",
            va='center', fontsize=10)

# Add title and labels
ax.set_title("Chi-Square Test Results", fontsize=14, weight='bold')
ax.set_xlabel("Chi-Square Value", fontsize=12)
ax.set_xlim(0, max(chi_square_values) + 10)

# Display the plot
plt.tight_layout()
plt.show()
```
